O basemap é ótima ferramenta para criar mapas usando o python de maneira simples. É uma extensão do matplotlib, logo ele possui todas as suas características para criar visualizações de dados, e adicionar projeções geográficas e alguns datasets para plotar linhas de costa, países, entre outros diretamente da biblioteca.

## 1) Desenhando o primeiro mapa.

```python
from mpl_toolkits.basemap import Basemap
import matplotlib.pyplot as plt
```

```python
map = Basemap()

map.drawcoastlines()

plt.show()
plt.savefig('test.png')
```

- 1: As duas primeiras linhas incluem a biblioteca Basemap e matplotlib. Ambas são necessárias.
- 2: O mapa é criado usando a classe Basemap, que possui muitas opções. Sem passar nenhuma opção, o mapa tem a projeção "Plate Carrée", centrada na latitude e longitude zero.
- 3: Depois de configurar o mapa, nós podemos desenhar o que queremos. Nesse caso, a camada com as linhas de costa, que já vem na biblioteca usando o método <code>drawcoastlines()</code>.
- 4: Finalmente, o mapa tem que ser mostrado ou salvo. Os métodos do matplotlib são usados pra isso. Neste exemplo, <code>plt.show</code> abre uma janela pra explorar o resultado. <code>plt.savefig('file_name')</code> salva o mapa como uma imagem.

Mudar a projeção é fácil, basta adicionar o argumento <code>projection</code>, <code>lat_0</code> e <code>lon_0</code> no construtor Basemap. 
Mesmo com a nova projeção, o mapa ainda é um pouco pobre, então vamos preencher os oceanos e continentes com algumas cores. Os métodos <code>fillcontinentes()</code> e <code>drawmapboundary()</code> farão isso:

```python
map = Basemap(projection='ortho', 
              lat_0=0, lon_0=0)

#Fill the globe with a blue color 
map.drawmapboundary(fill_color='aqua')
#Fill the continents with the land color
map.fillcontinents(color='coral',lake_color='aqua')

map.drawcoastlines()

plt.show()
```
